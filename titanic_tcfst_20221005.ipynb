```python
from urllib.request import urlretrieve
train_url = 'https://github.com/Elwing-Chou/ximen0926/raw/main/titanic/train.csv'
urlretrieve(train_url, 'train_titanic_tcfst_20221005.csv')
test_url = 'https://github.com/Elwing-Chou/ximen0926/raw/main/titanic/test.csv'
urlretrieve(test_url, 'test_titanic_tcfst_20221005.csv')
```

```python
import pandas as pd
train = pd.read_csv('train_titanic_tcfst_20221005.csv', encoding='utf-8')
predict = pd.read_csv('test_titanic_tcfst_20221005.csv', encoding='utf-8')
```

```python
print(len(train))
train
```

```python
predict
```

```python
combine = pd.concat([train, predict], axis=0, ignore_index=True)
combine
```

```python
combine = combine.drop(['PassengerId', 'Survived'], axis=1)
combine
```

```python
# Step1: 補缺失值
s = combine.isna().sum() #每個欄位有多少缺失值
s = s.sort_values(ascending=False)
s
```

```python
# Pandas篩選操作：DataFrame/Series[跟你資料筆數一樣多的True/False]
s[s > 0]
```

```python
# everything is an object
# 每一個東西都有形態

# Series.apply(轉換功能) ；※注意"轉換功能"後面不用加上()，因為pandas自動會將每列數值帶入函式。

def transfer_Name(n):
  # n = 'Kelly, Mr. James'
  n = n.split('.')[0].split(',')[-1]
  return n.strip()

combine['Name']= combine['Name'].apply(transfer_Name)
combine
```

```python
# value_counts() PS.不會計算NaN(Don’t include counts of NaN.)
dic = combine['Ticket'].value_counts()
dic
```

```python
def transfer_Ticket(t):
  if pd.isna(t): 
    return None
  else: 
    return dic[t]
combine['Ticket'] = combine['Ticket'].apply(transfer_Ticket)
combine
```

```python
def transfer_Cabin(c):
  if pd.isna(c): 
    return None
  else: 
    return c[0]
combine['Cabin'] = combine['Cabin'].apply(transfer_Cabin)
combine
```

```python
combine['Cabin'].value_counts()
```

```python
# 種類: Pclass、Name、Sex、Cabin、Embarked → 補最常出現的值
# 數值: Age、SibSp、Parch、Ticket、Fare → 補中位數
```

```python
# 種類
most = combine['Embarked'].value_counts().idxmax() #EX.大部分乘客是在S(Southampton南安普敦)搭船
combine['Embarked'] = combine['Embarked'].fillna(most)
combine
```

```python
# 確認'Embarked'已補完最常出現的值
s = combine.isna().sum() #每個欄位有多少缺失值
s = s.sort_values(ascending=False)
s
```

```python
# 數值
med = combine.median()
med

# ！！但要注意'Pclass'是「種類」資料！！
# 試想若大部分乘客是做"經濟艙"，但中間乘客是做"頭等艙"...補缺失值當然是要選擇"經濟艙"
#所以要把'Pclass'給drop，更正如下行程式碼。
```

```python
med = combine.median().drop('Pclass') 
med
```

```python
combine = combine.fillna(med)
combine
```

```python
# 確認數值資料已補完中位數
s = combine.isna().sum() #每個欄位有多少缺失值
s = s.sort_values(ascending=False)
s
```

```python
# Step2: One-Hot-Encoding
# 一定要做，但若太過稀少，可以去掉或改為其他(0, 0, 0)
combine['Name'].value_counts()
```

```python
counts = combine['Name'].value_counts()
f = counts > 50
counts[f]
```

```python
whitelist = counts[f].index
whitelist
```

```python
def transfer_Name(name):
  if name in whitelist: return name
  else: return None
combine['Name'] = combine['Name'].apply(transfer_Name)
combine
```

```python
# 種類: Pclass、Name、Sex、Cabin、Embarked → 補最常出現的值
# Pclass → EX."頭等艙"/"商務艙"/"經濟艙"，有大小關係，所以可做或不做One-Hot-Encoding。但測試過後，做One-Hot-Encoding後的模型較佳。
# Sex → 二值型資料，亦可做或不做One-Hot-Encoding。建議仍做One-Hot-Encoding。
# Name、Cabin、Embarked → 必做
pd.set_option('display.max_columns', 50)
combine = pd.get_dummies(combine) # Pandas會偵測資料型態，針對文字資料自動做One-Hot-Encoding
combine = pd.get_dummies(combine, columns=['Pclass']) # 針對'Pclass'補做One-Hot-Encoding
combine
```

```python
# Step3. 推斷一些額外資訊(萃取額外資訊)
combine["Family"] = combine["SibSp"] + combine["Parch"]
combine
```

```python
# iloc → index location(看第幾筆，index)
# loc → 列標籤
import numpy as np
x = combine.iloc[:len(train)]
x = np.array(x)
y = train['Survived']
```

```python
x_predict = combine.iloc[:len(train)]
x_predict = np.array(x_predict)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
clf = DecisionTreeClassifier(max_depth=4)
scores = cross_val_score(clf, x, y, cv=10, n_jobs=-1)
print('10:', scores)
print('average:', np.average(scores))

# 參數說明(https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_validate.html)
# estimator：模型，The object to use to fit the data.
# scoring:評分標準(metrics)，Strategy to evaluate the performance of the cross-validated model on the test set.
# cv: 資料切成幾分，Determines the cross-validation splitting strategy. 
# n_jobs: 平行運算，一次執行幾份工作。跟核心數量有關，四核心為一次處理四份工作；八核心為一次處理八份工作。Number of jobs to run in parallel. -1 means using all processors.
```

```python
# GridSearchCV貪婪演算法，找出最佳參數。
# 幫你把每組參數都做一次上面那格(cv)
from sklearn.model_selection import GridSearchCV
clf = DecisionTreeClassifier()
params = {'max_depth':range(3,20)}
search = GridSearchCV(clf, params, cv=10, n_jobs=-1)
search.fit(x, y)
```

```python
print('最佳分數', search.best_score_)
print('最佳參數', search.best_params_)
```
